## Composition control simulations
### Solve A=B ODE and plot composition dynamics for given initial conditions from time t0 to t1

Set your parameters

```python
%matplotlib inline
import numpy as np
#      x = 0, T1, 1, A1, 2, S1, 3, S2, 4, T2, 5, A2, 6, C1, 7, C2
#      P = 0, beta_S1, 1, l_S1, 2, K_S1, 3, kb, 4, beta_S2, 5, l_S2, 6,
#      K_S2, 7, beta_lac, 8, l_lac, 9, K_lac, 10, beta_tet, 11, l_tet, 12,
#      K_tet, 13, kc, 14, C_max, 15, dc, 16, xx, 17, I, 18, xx, 19, atc, 20,K_tox
P = np.zeros(24)
P[0] = 6
P[1] = 2e-3
P[2] = 430
P[3] = 4
P[4] = 6
P[5] = 2e-3
P[6] = 430
P[7] = 19.8e-3
P[8] = 1.5e-3
P[9] = 1.4e5
P[10] = 14.4e-3
P[11] = 2.1e-4
P[12] = 13
P[13] = 0.6
P[14] = 5500
P[15] = 0.8
P[16] = np.Inf
P[17] = 1e6
P[18] = np.Inf 
P[19] = 324
P[20] = 1
P[21] = 0.001
P[22] = 1.5
P[23] = 0
```

```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt 
from reed_ode import reed_ode
ODE_size = 8

#  Initial seeding (make your own choice)
init_cond = [ [100,100],[100,110],[200,200],[200,220],[300,330],[300,300]]    
init_cond = [ [1000,500],[500,1000],[200,200],[100,100],[300,1500],[2000,400]]    
# init_cond = [[100,100],[500,600]]

t0 = 0
t1 = 80
timepoints = np.linspace(t0,t1,100) 

x0=np.zeros(ODE_size)

for i in range(len(init_cond)):
    x0[6] = init_cond[i][0]
    x0[7] = init_cond[i][1]
    outs = solve_ivp(fun = lambda t, x: reed_ode(t, x, P) , t_span = (t0,t1), y0 = x0, t_eval = timepoints)
    t = outs.t
    y = outs.y[:,]
    comp = [i/j for i,j in zip(y[6,:],y[7,:])]
    legend = 'Initial '+ str(x0[6]) + ' : ' + str(x0[7]) 
    plt.plot(timepoints, comp, label = legend)
plt.legend()
plt.show()

```

### Solve A=B ODE with disturbance to initial condition after time t1. Plots composition control for given initial condition vector

```python
disturb = 1000
tn = 160
timepoints2 = np.linspace(t1,tn,100)
for i in range(len(init_cond)):
    x0[6] = init_cond[i][0]
    x0[7] = init_cond[i][1]
    outs = solve_ivp(fun = lambda t, x: reed_ode(t, x, P), t_span = (t0,t1), y0 = x0, t_eval = timepoints)
    t_one = outs.t
    y_one = outs.y[:,]

 # Disturbance to initial condition
    x0n = [i for i in y_one[:,-1]]
    x0n[6] += disturb
    x0n[7] += 0.5*disturb
    outs1 = solve_ivp(fun = lambda t, x: reed_ode(t, x, P), t_span = (t1,tn), y0 = x0n, t_eval = timepoints2)
    tf = outs1.t
    yf = outs1.y[:,]
    x_result = [np.concatenate((i,j)) for i,j in zip(y_one, yf)]
    time_final = np.concatenate((t_one, tf))
    y_result_C1 = list(x_result[6])
    y_result_C2 = list(x_result[7])
#     plt.plot(time_final, y_result_C2)
    # Total population
    legend1 = 'Initial '+ str(x0[6]) + ' : ' + str(x0[7]) + ', and d = ' + str(disturb)
    comp = [i/j for i,j in zip(y_result_C1,y_result_C2)]
    plt.plot(time_final,comp, label = legend1)
    
plt.legend()
plt.show()
```
